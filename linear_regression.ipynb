```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error


learning_rate = 0.01
num_iterations = 10000
convergence_threshold = 0.00001

data = pd.read_csv('C:\\Users\\dell\\C_programs\\data.txt')

X = data['X']
y = data['Y']

X = (X - X.mean()) / X.std()

num_samples = len(X)
theta0 = 0 
theta1 = 0 
prev_cost = float('inf')

for iteration in range(num_iterations):
    y_pred = theta0 + theta1 * X
    gradient0 = -2 * (y - y_pred).mean()
    gradient1 = -2 * (X * (y - y_pred)).mean()
    theta0 -= learning_rate * gradient0
    theta1 -= learning_rate * gradient1
    cost = mean_squared_error(y, y_pred)    
    if abs(prev_cost - cost) < convergence_threshold:
        print(f"Converged after {iteration} iterations.")
        break
    
    prev_cost = cost

y_pred = theta0 + theta1 * X
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)

plt.scatter(X, y, label="Actual Data")
plt.plot(X, y_pred, color='red', label="Linear Regression Line")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Linear Regression with Bias Term")
plt.legend()
plt.show()

print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```
